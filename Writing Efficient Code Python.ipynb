## Problem 1(a).  Unionized workers.


ðŸŽ¯ Create a pandas data frame to hold the CPS data.  
ðŸŽ¯ Write a function that computes the average hourly wage of a unionized worker by using **control flow** to iterate through each row of the data frame.  There must be a loop in this function.

````python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
````

````python
cps = pd.read_csv("cps.csv")
cpsframe = DataFrame(cps)
cpsframe
````

````python
def loopWage(row):
    wage = 0
    count = 0
    for i, row in cpsframe.iterrows():
        if row['union'] == "Union":
            wage += row['wage']
            count += 1
    meanwage = wage/count
    
    return meanwage

loopWage(cpsframe)
````

## Problem 1(b).  More unionized workers.

ðŸŽ¯ Write a function that takes a row of a data frame as input.

Use tab completion or documentation for pandas series objects to find some useful functions!  Numpy rules, too.

````python
def ApplyWage(row):
    aWage = []
    if row['union'] == "Union":
        aWage.append(row['wage'])
    else:
        aWage.append(np.nan)
    aWage = np.array(aWage)
    return aWage
````

````python
def ApplyFunc(row):
    unionWage = cpsframe.apply(ApplyWage, axis=1)
    meanWage = np.nanmean(unionWage)
    return meanWage
ApplyFunc(cpsframe)
````
